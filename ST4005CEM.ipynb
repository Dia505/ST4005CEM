```python
# Numpy aids in working with arrays and manipulating them
import numpy as np
# Pandas is used for data cleaning and preparation
import pandas as pd
# Matplotlib is for data plotting and visualization
import matplotlib.pyplot as plt


# To read the csv file and convert it into a data frame
# index_col = 0 removes the index
dataf = pd.read_csv("airlines.csv", index_col = 0)
# To display the data frame
dataf
```

```python
# To check if any null values are present 
null = dataf.isnull().sum()
null
# There are 83 null values in column "Arrival Delay in Minutes"
```

```python
# The null values are filled with the preceding non-missing value using the function fillna()
# Here the method ffill is used where the first "f" stands for forward
filled_dataf = dataf.fillna(method = "ffill")
filled_dataf
```

```python
# A sample dataframe is created by randomly selecting 20 rows from the filled_dataf dataframe
df_sample = filled_dataf.sample(20)
df_sample
```

```python
# Bar graph to compare inflight entertainment in eco, business and eco plus class
# Column "Class" is kept in variable "flight"
flight = df_sample["Class"]
# Column "Inflight entertainment" is kept in variable "entertainment"
entertainment = df_sample["Inflight entertainment"]
# Creation of bar graph
plt.bar(flight,entertainment,width = 0.5)
# Data of flight displayed in x-axis
plt.xlabel("Flight class")
# Data of entertainment displayed in y-axis
plt.ylabel("Entertainment rating")
plt.title("Inflight entertainment in different flight class")
plt.show()
```

```python
# Histogram showing the variation of ages of passengers
# Column "Age" is kept in variable "age"
age = df_sample["Age"]
# Creation of histogram
plt.hist(age, color = "purple")
plt.xlabel("Passenger age")
plt.ylabel("Flight frequency")
plt.title("Passenger age distribution")
plt.show()
```

```python
# Histogram showing the frequency of flight distances travelled
distance = df_sample["Flight Distance"]
plt.hist(distance, color = "green")
plt.xlabel("Flight distance")
plt.ylabel("Frequency")
plt.title("Distribution of flight distances")
plt.show()
```

```python
# Pie chart that shows the average cleanliness rating received by eco, eco plus and business class flights
# The dataframe is grouped by "Class" column using groupby() and the average values from "Cleanliness" column is calculated using mean()
average_cleanliness = df_sample.groupby("Class")["Cleanliness"].mean()
# Creation of pie chart
# average_cleanliness.index refers to Eco, Eco Plus and Business
# autopct is used to display percentage on each zone
plt.pie(average_cleanliness, labels=average_cleanliness.index, autopct="%1.1f%%")
plt.title("Cleanliness in different flight classes")
plt.show()
```

```python
# Line plot that represents the relation between Flight distance and On-board services
# X-axis: Flight Distance
# Y-axis: On-board services ratings
plt.plot(df_sample["Flight Distance"][0:5],df_sample["On-board service"][0:5])
plt.show()
```

```python
# Regression
# Statsmodels is used for linear regression
import statsmodels.api as sm

# Columns "Arrival Delay in Minutes" and "Departure Delay in Minutes" are independent variables
# add_constant() adds constant term to the dataframe X
X = sm.add_constant(filled_dataf[["Arrival Delay in Minutes","Departure Delay in Minutes"]])
# "Departure/Arrival time convenient" is the dependent variable 
Y = filled_dataf["Departure/Arrival time convenient"]
# sm.OLS() fits the linear regression model or estimates the model's parameters
model = sm.OLS(Y,X).fit()

# A new dataframe reg_df is created with independent variables as columns
newData = {
    "Arrival Delay in Minutes":filled_dataf["Arrival Delay in Minutes"][0:15],
    "Departure Delay in Minutes":filled_dataf["Departure Delay in Minutes"][0:15]
}
reg_df = pd.DataFrame(newData)
reg_df
```

```python
# reg_df dataframe is used to make predictions of dependent variable  
new_X = sm.add_constant(reg_df)
predictions = model.predict(new_X)
predictions
```

```python
# To evaluate the result, Mean Absolute Error is calculated
# A data series is created with the actual values 
actual_data = filled_dataf["Departure/Arrival time convenient"][0:15]
y_true = pd.Series(actual_data)
y_true
```

```python
# The absolute difference is calculated via abs() of numpy library
absolute_diff = np.abs(predictions - y_true)
# mean() of numpy calulates MAE
mae = np.mean(absolute_diff)
mae
```

```python
# Since the Mean Absolute Error is quite greater than 0, the predicted data is not accurate
# A separate sample dataframe is created by randomly selecting 15 rows from the filled_dataf dataframe
dtf_sample = filled_dataf.sample(15)
dtf_sample
```

```python
# Creation of another regression model using the sample data frame "dtf_sample1"
import statsmodels.api as sm

X1 = sm.add_constant(dtf_sample[["Arrival Delay in Minutes","Departure Delay in Minutes"]])
Y1 = dtf_sample["Departure/Arrival time convenient"]
s_model = sm.OLS(Y1,X1).fit()

s_Data = {
    "Arrival Delay in Minutes":dtf_sample["Arrival Delay in Minutes"],
    "Departure Delay in Minutes":dtf_sample["Departure Delay in Minutes"]
}
sample_reg = pd.DataFrame(s_Data)
sample_reg
```

```python
sample_X = sm.add_constant(sample_reg)
sample_predictions = s_model.predict(sample_X)
sample_predictions
```

```python
actual_sample = dtf_sample["Departure/Arrival time convenient"]
y_sample = pd.Series(actual_sample)
y_sample
```

```python
s_absolute_diff = np.abs(sample_predictions - y_sample)
sample_mae = np.mean(s_absolute_diff)
sample_mae
```

```python
# The Mean Absolute Error is closer to 0, which means the predictions are more accurate
# The intercept of s_model
intercept = s_model.params[0]
intercept
```

```python
# The slope of s_model
slope = s_model.params[1]
slope
```

```python
# Visualization of the predicted data in comparison to the actual data
# Dataframe result_df created with column "Actual" representing actual values and "Predicted" representing predicted values
result_df = pd.DataFrame({"Actual": y_sample, "Predicted": sample_predictions})
# The dataframe is sorted by the actual values for better visualization
result_df.sort_values(by="Actual", inplace=True)

# Scatter plot of actual and predicted values
plt.scatter(result_df["Actual"], result_df["Predicted"], label="Predicted vs. Actual")

# Line plot represents the actual values i.e., a condition of perfect prediction
plt.plot(result_df["Actual"], result_df["Actual"], color="red", label="Perfect Prediction")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Predicted vs. Actual Values")
plt.show()

# The figure shows a couple of data points nearby the red line, representing quite accurate predictions.
# The data points that are farther away from the line represent inaccurate predictions.
```

```python
# Bar graph illustrating the difference between actual and predicted values
# Creation of two bar graphs overlapping each other to represent the difference
plt.figure(figsize=(15, 6)) 

# Bar graph of predicted values
plt.bar(result_df.index, result_df["Predicted"], color="green", alpha = 0.7, label="Predicted", width = 400)
# Bar graph of actual values
plt.bar(result_df.index, result_df["Actual"], color="orange", alpha=0.7, label="Actual", width = 400)
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.title("Predicted vs. Actual Values")
plt.show()

# Most of the green and orange rectangular bars are quite close in height, which shows the similarities between the predicted and actual data
```

```python
# Histogram of differences calculated from actual and predicted values
difference = result_df["Actual"] - result_df["Predicted"]
plt.hist(difference, color = "brown")
plt.xlabel("Difference")
plt.ylabel("Frequency")
plt.title("Differences from predicted and actual values")
plt.show()

# The differences are fairly symmetrical in some areas, with a few inaccuracies as seen below. 
```

```python
# Correlation matrix of columns On-board service, Cleanliness, Inflight entertainment and Flight Distance of data sample "dtf_sample"
# Correlation matrix created to see the relation between the columns
corr_columns = ["On-board service","Cleanliness","Inflight entertainment","Flight Distance"]
corr_matrix = dtf_sample[corr_columns].corr(method = "pearson")
corr_matrix.reset_index(drop = True)
corr_matrix

# Positive correlation coefficients explain that when one variable increases, the other variables tend to increase too
# Negative correlation coefficients explain that when one variable increases, the other variables tend to decrease
# Correlation coefficients with value greater than 0.3 show moderately strong relations, while the values less than 0.3 show weak relations
# Flight Distance is not really related to the other variables
# Inflight entertainment has moderate connections with On-board service and Cleanliness
# It shows that On-board service and Inflight entertainment ratings decrease with longer flight distances
```
